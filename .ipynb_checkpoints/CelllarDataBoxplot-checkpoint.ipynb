```python
# Dependencies and Setup
import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
import csv

# Import API key
from config import gkey
```

```python
# pull in cell tower csv.
cell_df = pd.read_csv("Resources/OCID-diff-cell-export-2019-09-23-T000000.csv")

# limit mcc code to only view US cell towers. mcc 310 thru 316
cell_df = cell_df[(cell_df['mcc'] >= 310) & (cell_df['mcc'] <= 316)]

cell_df.head()
```

```python
df = cell_df[(cell_df['radio'] == "LTE")]
df.head()
```

```python
df2 = df.loc[df["net"].isin([120])]
```

```python
import seaborn as sns
ax = sns.boxplot(x="net", y="cell", data=df2)
ax = sns.swarmplot(x="net", y="cell", data=df2)    
plt.xlabel('Carrier')
plt.ylabel('Cell')
plt.show()
```
